```python
applications_df = pd.read_csv("Documents/applications.csv")
industries_df = pd.read_csv("Documents/industries.csv")
```

```python
cleaned_applications_df = applications_df.drop_duplicates(subset=["applicant_id"], ignore_index=True)
cleaned_applications_df
```

```python
cleaned_applications_df["External Rating"].fillna(0)
cleaned_applications_df
```

```python
cleaned_applications_df["Education level"].fillna("Середня")
cleaned_applications_df
```

```python
merged_df = pd.merge(left=cleaned_applications_df, right=industries_df, on='Industry',how='left')
merged_df
```

```python
filter1 = ((merged_df['Age'] >= 35) & (merged_df['Age'] <= 55)) * 20
filter1
#filter by age
```

```python
merged_df['Applied at'] = pd.to_datetime(merged_df['Applied at'], format='mixed')
merged_df.dtypes
```

```python
merged_df['Day of the Week'] = merged_df['Applied at'].dt.day_name()
merged_df
```

```python
merged_df['Day of the Week'] = merged_df['Applied at'].dt.day_name()
merged_df
```

```python
filter2 = (~(merged_df['Day of the Week'] == 'Saturday') | (merged_df['Day of the Week'] == 'Sunday')) * 20
filter2
#not weekend
```

```python
filter3 = (merged_df["Marital status"] == "Married") * 20
filter3
#if married
```

```python
filter4 = (merged_df["Location"] == "Київ чи область") * 10
filter4
#if from Kyiv
```

```python
filter5 = (merged_df["External Rating"] >= 7 ) * 20
filter5
#if 'External Rating' >=7
```

```python
filter6 = (merged_df["External Rating"] <= 2 ) * (-20)
filter6
#if 'External Rating' <=2
```

```python
merged_df['Rating'] = filter1 + filter2 + filter3 + filter4 + filter5 + filter6 + merged_df['Score']
merged_df
```

```python
merged_df['Rating'] = merged_df.apply(lambda row: 0 if pd.isnull(row['Amount']) or row['External Rating'] == 0 else row['Rating'], axis=1)
merged_df
```

```python
final_df = merged_df[(merged_df['Rating'] <= 100) & (merged_df['Rating'] >= 0)].reset_index(drop=True)
final_df
```

```python
final_df['Week Number'] = final_df['Applied at'].dt.isocalendar().week
final_df
```

```python
pivot_table = final_df.pivot_table(index='Week Number', values='Rating', aggfunc='mean').round(2)
pivot_table
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(pivot_table.index, pivot_table['Rating'], color = "green")
plt.title('Середній рейтинг прийнятих заявок за тижнем')
plt.xlabel('Тиждень')
plt.ylabel('Середній рейтинг')
plt.grid(True)
plt.show()
```
